# Regressão linear múltipla

A regressão linear múltipla é semelhante à regressão linear simples, porém com mais de uma variável independente, significando que a variável a ser predita é baseada em 2 ou mais variáveis de entrada.

Matematicamente temos:

\begin{equation}
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \ldots + \beta_nx_n
\end{equation}

onde $y$ é a variável dependente, $\beta_0$, a constante, $\beta_1$ a inclinação em relação à variável $x_1$, $\beta_2$ a inclinação em relação à variável $x_2$ e assim por diante.

Para que a regressão linear múltipla traga bons resultados é preciso que:

1. Existe uma relação linear entre a variável dependente e as variáveis independentes.
2. As variáveis independentes não tenham alta correlaçãno entre si.
3. O tamanho do residuo não depende do tamanho do valor de entrada (homocedasticidade)
4. Os resíduos apresentam distribuição normal

# Regressão linear de polinômios

Caso o modelo a ser estimado tenha o formato de um polinômio:

\begin{equation}
y = \theta_0 + \theta_1x + \theta_2x^2 + \ldots \theta_nx^n
\end{equation}

é possível utilizar a função `polyfit` da biblioteca `numpy`.

```python
# https://cn.ect.ufrn.br/index.php?r=conteudo%2Fmmq-rpolin  
from numpy.polynomial import polynomial
import matplotlib.pyplot as plt

grau = 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([2.1, 7.7, 13.6 ,27.2 ,40.9 ,61.1])
#res = np.polyfit(x,y,grau)
res = polynomial.polyfit(x,y,grau)
print(res)

th0 = res[0]
th1 = res[1]
th2 = res[2]

yest = th2*x**2 + th1*x + th0
plt.scatter(x, y)
plt.plot(x, yest, 'r')

# também é possível utilizar a função polyval:
y2 = polynomial.polyval( x, res)
print(yest, '\n', y2)
```

# Underfitting x Overfitting

https://edisciplinas.usp.br/pluginfile.php/5775889/mod_resource/content/2/Aula03_Regression.pdf

O **viés** representa o erro do valor esperado em relação ao valor predito pelo modelo:

\begin{equation}
Bias = E[\hat{y} - y]
\end{equation}

onde $\hat{y}$ é o valor predito pelo modelo e $y$ o valor observado.

Já a **variância** mede a distância de cada estimativa em relação à sua média:

\begin{equation}
Var = E[\hat{y^2}] - E[\hat{y}]^2
\end{equation}

![vies](./figuras/viesevariancia.png)

Existe um trade-off entre a complexidade do modelo (grau do polinômio) e o erro total (viés + variância).
Em geral, quanto mais simples o modelo, maior o seu viés, e quanto mais complexo o modelo, maior a sua variância.

![overfitting](./figuras/overfitting.png)

Geralmente existem um ponto ótimo da complexidade do modelo que minimiza o erro total.

![complexidade](./figuras/complexidade_modelo.png)


## Exemplo

Considere os pontos a seguir:

$x = [0, 0.18, 0.25, 0.4, 0.45, 0.55, 0.63, 0.75, 0.85, 1]$

$y = [0.3, 0.8, 1, 0.95, 0.25, 0.3, -0.9, -0.7, -0.8, 0.35]$

Em seguida, plote o gráfico dos modelos aproximando por um polinômio de grau 1, 3, 6 e 8

```python
from numpy.polynomial import polynomial
import numpy as np
import matplotlib.pyplot as plt

x = [0, 0.18, 0.25, 0.4, 0.45, 0.55, 0.63, 0.75, 0.85, 1]
y = [0.3, 0.8, 1, 0.95, 0.25, 0.3, -0.9, -0.7, -0.8, 0.35]
X = np.linspace(0,1,100)

cor = ['cyan', 'red', 'lime', 'magenta'] 
estilo = ['solid', 'dashed', 'solid', 'dotted']
plt.figure(figsize=(8, 5), dpi=150)

plt.scatter(x,y, c='gray')

for i, grau in enumerate([1,3, 6, 8]):
    res = polynomial.polyfit(x,y,grau)

    yest = polynomial.polyval(X, res)
    plt.plot(X, yest, cor[i], ls=estilo[i], lw=0.7)

plt.legend(['observações','grau 1', 'grau 3', 'grau 6', 'grau 8'])
```

# R2 ajustado

Para múltiplas variáveis, é preferível utilizar o parâmetro $R^2$ ajustado, que leva em consideração o quanto uma nova variável ajuda a explicar os dados experimentais, penalizando novas variáveis irrelevantes, que contribuem pouco:

\begin{equation}
R^2_{adj} = 1 - \dfrac{(n-1)}{(n-1)-k}(1-R^2)
\end{equation}

ou seja, quando maior o número de regressores, maior a penalidade.

```python
def r2adj(r2, k, n):
    return 1 - (n-1)*(1-r2)/(n-1-k)
```

```python
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import polynomial

from sklearn.metrics import r2_score


x = [0, 0.18, 0.25, 0.4, 0.45, 0.55, 0.63, 0.75, 0.85, 1]
y = [0.3, 0.8, 1, 0.95, 0.25, 0.3, -0.9, -0.7, -0.8, 0.35]
X = np.linspace(0,1,100)

df = pd.DataFrame(columns=['y', 'x'])
df['x'] = x
df['y'] = y

cor = ['cyan', 'red', 'lime', 'magenta'] 
estilo = ['solid', 'dashed', 'solid', 'dotted']
plt.figure(figsize=(8, 5), dpi=150)

plt.scatter(x,y, c='gray')

for i, grau in enumerate([1,3, 6, 8]):
    coeficientes = polynomial.polyfit(x,y,grau)
    
    ypredito = polynomial.polyval(x, coeficientes)
    
    r2 = r2_score(y, ypredito)
    r2ajustado = r2adj(r2, grau, len(x))

    print(f'grau: {grau}, r2: {r2}, r2 ajustado: {r2ajustado}')

    yest = polynomial.polyval(X, coeficientes)
    plt.plot(X, yest, cor[i], ls=estilo[i], lw=0.7)

plt.legend(['observações','grau 1', 'grau 3', 'grau 6', 'grau 8'])
```
