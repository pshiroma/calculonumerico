# Erro



$\text{valor verdadeiro} = \text{aproximação} + \text{erro}$

Onde:

* valor verdadeiro: Valor real, conhecido a partir da solução analítica
* aproximação: Valor obtido por um método numérico
* erro: diferença entre o valor verdadeiro e a aproximação numérica

O **erro relativo percentual verdadeiro** mede o erro cometido por uma **aproximação** realizada por algum método numérico, normalizado pela ordem de grandeza do valor verdadeiro. 

$\text{erro relativo percentual verdadeiro} = \dfrac{\text{valor verdadeiro} - \text{aproximação}}{\text{valor verdadeiro}} 100\%$

Porém, na maioria dos casos não é possível obter o valor do valor verdadeiro.

$\text{erro relativo percentual aproximado} = \dfrac{\text{erro aproximado}}{\text{aproximação}}100\%$

Uma forma mais apropriada é calcular o **erro relativo percentual aproximado**: 

$\text{erro relativo percentual aproximado} = \dfrac{\text{aproximação atual} - \text{aproximação prévia}}{\text{aproximação atual}}100\%$

# Exemplo: Derivada

O conceito de derivada está relacionado à taxa de variação instantânea de uma função.
Uma definição da derivada de uma função $f(x)$ é:

\begin{equation}
f'(x) = \lim\limits_{\Delta x \rightarrow 0} \dfrac{f(x + \Delta x) - f(x)}{\Delta x}
\end{equation}

\begin{equation} \label{eq:taylorprogressiva}

\end{equation}

\begin{equation}

\end{equation}

\begin{equation}
