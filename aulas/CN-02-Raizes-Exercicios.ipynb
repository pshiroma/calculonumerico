## Exercício 6

primos.drawio.svg

Suponha que você tenha um tanque cilíndrico deitado e queira fazer marcações na parte externa.
Cada marcação deve corresponder ao volume interno preenchido pelo líquido, conforme figura acima.

Observe que, devido ao formato do cilindro, o ângulo onde devem ser feito as marcações não tem um espaçamento uniforme.

De fato, é possível mostrar que o volume se relaciona com o ângulo a partir da seguinte equação:

\begin{equation} 
v(\theta) = lr^2(\theta -\dfrac{1}{2}\sin 2\theta)
\end{equation}

onde $l$ é comprimento do tubo, $\theta$ varia de 0 até $\pi$ e $r$ é o raio do cilindro.

Determine o ângulo (em radianos) para preencher um volume de 1$l$, 5$l$ e 10$l$ usando um tubo de comprimento $l = 5m$ e raio $r = 2m$

```python
def f(theta):
    l = 5
    r = 2
    vdes = 5
    v = l*r**2*(theta - np.sin(2*theta)/2)
    return vdes - v
```

```python
import numpy as np
import scipy.optimize as opt
sol = opt.root_scalar(f, method='bisect', bracket=[0,2])
np.degrees(sol.root)
```
